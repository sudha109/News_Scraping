```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import MoveTargetOutOfBoundsException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\eshan\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
url="https://www.deccanherald.com/archives"
driver.get(url)
driver.maximize_window()
```

```python
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
```

```python
def Details():
    links = [link.get_attribute('href') for link in driver.find_elements_by_xpath("//div[@class='group'][2]//div//div//ul//li//a")]
    for link in links:
        driver.get(link)
        time.sleep(3)
        driver.implicitly_wait(10)
    
        try:
            date=driver.find_element_by_xpath("//div[@class='article-author__time']//ul[@class='crud-items']//li[1]")
            dat=date.text
            Date.append(dat)
        except NoSuchElementException:
            Date.append('not available')
    
        try:
            author=driver.find_element_by_xpath("//div[@class='article-author__time']//span[1]")
            aut=author.text
            Author.append(aut)
        except NoSuchElementException:
            Author.append('not available')
    
        try:
            vertical=driver.find_element_by_xpath("//div[@class='section-container']//div//ul//li[2]//a")
            ver=vertical.text
            Vertical.append(ver)
        except NoSuchElementException:
            Vertical.append('not available')
        
        try:
            headline=driver.find_element_by_xpath("//div[@class='section-container']//div//ul//li[@class='last']")
            head=headline.text
            Headline.append(head)
        except NoSuchElementException:
            Headline.append('not available')
        
        try:
            description=driver.find_element_by_xpath("//div[@class='field field-name-body field-type-text-with-summary field-label-hidden']//div")
            des=description.text
            Description.append(des)  
        except NoSuchElementException:
            Description.append('not available')
            
        
            
    
    
```

```python
def dates():
    date=driver.find_element_by_xpath("//div[@class='form-group']//a")
    driver.implicitly_wait(10)
    ActionChains(driver).move_to_element(date).click(date).perform()

    year=driver.find_element_by_xpath("//div[@class='numInputWrapper']//span[@class='arrowDown']")
    for x in range(4):
        year.click()
    
    month=driver.find_element_by_xpath("//span[@class='flatpickr-prev-month']")
    for x in range(6):
        month.click()
    
```

```python
dates()
days=driver.find_elements_by_xpath("//div[@class='dayContainer']//span[@class='flatpickr-day ']")
number_of_days = len(days)
for x in range(29):
    day=driver.find_elements_by_xpath("//div[@class='dayContainer']//span[@class='flatpickr-day ']")[x]
    day.click()
    driver.implicitly_wait(10)
    try:       
        submit=driver.find_element_by_xpath("//a[@class='sanspro-semib datepicker-submit sq-horoscope-bg__button--black-bg sq-horoscope-bg__button']")
        submit.click()
    except StaleElementReferenceException:
        pass
    
      
    Details()
    
    try:
        driver.get(url)
    except StaleElementReferenceException:
        print('Refresh!')
        driver.refresh()
        driver.get(url)
        time.sleep(3)
    try:
        dates()
    except ElementNotInteractableException:
        driver.implicitly_wait(10)
        dates()
    
    
     
```

```python
news=pd.DataFrame({})
news['Date']=Date
news['Author']=Author
news['Vertical']=Vertical
news['Headline']=Headline
news['Description']=Description
```
